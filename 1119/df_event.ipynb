```python
# Load the match data JSON
data = pd.read_json('2043670_match_data.json', lines=True)

# Extract player details from the 'players' field
player_rows = []
if 'players' in data.columns:
    for player_list in data['players']:
        if isinstance(player_list, list):
            for player in player_list:
                if isinstance(player, dict):
                    player_id = player.get('id')
                    team_id = player.get('team_id')
                    first_name = player.get('first_name')
                    last_name = player.get('last_name')
                    short_name = player.get('short_name')
                    player_rows.append({
                        'player_id': player_id,
                        'team_id': team_id,
                        'first_name': first_name,
                        'last_name': last_name,
                        'short_name': short_name
                    })

# Create a DataFrame of all unique players by player_id
player_df = pd.DataFrame(player_rows).drop_duplicates(subset=['player_id']).sort_values('player_id').reset_index(drop=True)
player_df
```

```python
# Export the player table to an Excel file
player_df.to_excel('player_table.xlsx', index=False)
print('Exported player table to player_table.xlsx')
```
